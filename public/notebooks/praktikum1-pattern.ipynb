```python
import math
# Koordinat titik p1 dan p2
x1, y1 = 1, 5
x2, y2 = 4, 7
```

```python
# Menghitung panjang sisi-sisi segitiga
sisi_a = abs(x2 - x1)
sisi_b = abs(y2 - y1)
```

```python
# Menggunakan rumus Pythagoras untuk menghitung jarak
jarak = math.sqrt(sisi_a**2 + sisi_b**2)
print("Jarak antara titik p1 dan p2 adalah:", jarak)
```

```python
import numpy as np
# Sekumpulan data statistik
data = np.array([2.5, 8.2, -1.1, -0.2, 1.5])
```

```python
# Menghitung rerata (mean)
rerata = np.mean(data)
```

```python
# Menghitung variansi
variansi = np.var(data)
```

```python
# Menghitung standar deviasi
standar_deviasi = np.std(data)
```

```python
print("Rerata:", rerata)
print("Variansi:", variansi)
print("Standar Deviasi:", standar_deviasi)
```

**KASUS 4**

```python
import math
def akar2(a, b, c):
    # Menghitung diskriminan
    d = b**2 - 4*a*c

    # Menghitung akar-akar
    x1 = (-b + math.sqrt(d)) / (2*a)
    x2 = (-b - math.sqrt(d)) / (2*a)
    return x1, x2
```

```python
# Menghitung akar-akar persamaan kuadrat dengan a=1, b=-1, dan c=-12
x1, x2 = akar2(1, -1, -12)
```

```python
# Mencetak akar-akar
print("Akar pertama:", x1)
print("Akar kedua:", x2)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Membuat array x dari 0 hingga 2*pi dengan peningkatan 0.01
x = np.arange(0, 2*np.pi, 0.01)
```

```python
# Menghitung nilai y = sin(x)
y = np.sin(x)
```

```python
# Menampilkan grafik
plt.plot(x, y)
plt.title('Grafik dari y = sin(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

**KASUS 6**

```python
import numpy as np
import matplotlib.pyplot as plt

# Membuat array x dari 0 hingga 2*pi dengan peningkatan pi/100
x = np.arange(0, 2*np.pi, np.pi/100)
```

```python
# Menghitung nilai y untuk setiap fungsi kosinus
y1 = 2 * np.cos(x)
y2 = np.cos(x)
y3 = 0.5 * np.cos(x)
```

```python
# Menampilkan grafik
plt.plot(x, y1, '--', label='2*cos(x)')
plt.plot(x, y2, '-', label='cos(x)')
plt.plot(x, y3, ':', label='0.5*cos(x)')
plt.xlabel(r'$0 \leq x \leq 2\pi$')
plt.ylabel('Nilai fungsi y(x)')
plt.legend()
plt.title('Grafik Fungsi-fungsi Cosinus')
plt.axis([0, 2*np.pi, -3, 3])
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Membuat array x dari 0 hingga 2*pi dengan peningkatan 0.01
x = np.arange(0, 2*np.pi, 0.01)
```

```python
# Menghitung nilai y untuk setiap fungsi
y1 = np.sin(2*x)
y2 = np.cos(2*x)
y3 = 0.5 * np.cos(5*x)
y4 = 2 * np.sin(3*x)
```

```python
# Menampilkan grafik dalam subplot
plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title('sin(2x)')

plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title('cos(2x)')

plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title('0.5*cos(5x)')

plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title('2*sin(3x)')

plt.tight_layout()
plt.show()
```
