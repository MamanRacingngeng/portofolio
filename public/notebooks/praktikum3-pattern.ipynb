```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import kurtosis, skew
```

# **Masukkan data:**

```python
import pandas as pd
```

```python
# Muat dataset
data = pd.read_csv('IRIS.csv')
```

# **Ekstraksi Fitur dengan Ordo Pertama:**

```python
# Ekstraksi fitur ordo pertama untuk setiap kolom numerik
numeric_features = data.select_dtypes(include=np.number).columns

for feature in numeric_features:
    mean_value = np.mean(data[feature])
    variance_value = np.var(data[feature])
    std_deviation_value = np.std(data[feature])
    kurtosis_value = kurtosis(data[feature])
    skewness_value = skew(data[feature])

    print(f"Hasil Ekstraksi Fitur untuk {feature}:")
    print("Mean:", mean_value)
    print("Variance:", variance_value)
    print("Standard Deviation:", std_deviation_value)
    print("Kurtosis:", kurtosis_value)
    print("Skewness:", skewness_value)
    print("\n")
```

# **Visualisasikan dalam bentuk grafik bar dan plot**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Muat dataset
data = pd.read_csv('IRIS.csv')

# Ekstraksi fitur dan perhitungan statistik
numeric_features = data.select_dtypes(include=np.number).columns
features = ['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width']  # Ganti dengan nama fitur yang sesuai

mean_values = []
min_values = []
max_values = []

for feature in numeric_features:
    mean_values.append(np.mean(data[feature]))
    min_values.append(np.min(data[feature]))
    max_values.append(np.max(data[feature]))

# Visualisasi dalam bentuk bar plot
plt.figure(figsize=(10, 6))
plt.bar(features, mean_values, color='skyblue', label='Mean')
plt.bar(features, min_values, color='orange', label='Min', bottom=mean_values)  # Min di bawah mean
plt.bar(features, max_values, color='green', label='Max', bottom=[m + n for m, n in zip(mean_values, min_values)])  # Max di atas min dan mean
plt.title('Fitur Statistik')
plt.xlabel('Fitur')
plt.ylabel('Nilai')
plt.legend()
plt.show()

# Visualisasi data asli dalam bentuk histogram (opsional)
plt.figure(figsize=(10, 6))
for feature, name in zip(numeric_features, features):
    plt.hist(data[feature], bins=20, alpha=0.5, label=name)
plt.title('Distribusi Fitur')
plt.xlabel('Nilai')
plt.ylabel('Frekuensi')
plt.legend()
plt.show()
```

# **KASUS 2**

# **Import Library yang dibutuhkan:**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import kurtosis, skew
```

# **Masukkan data:**

```python
# Data array tinggi maskot bangunan negara (dalam meter)
maskot_bangunan = ['Puncak Jaya', 'Gunung Karinci', 'Gunung Rinjani', 'Gunung Semeru', 'Gunung Sanggar']
tinggi = np.array([4884, 3805, 3726, 3676, 3564])
```

# **Ekstraksi Fitur dengan Ordo Pertama:**

```python
# Ekstraksi fitur ordo pertama
mean_value = np.mean(tinggi)
variance_value = np.var(tinggi)
std_deviation_value = np.std(tinggi)
kurtosis_value = kurtosis(tinggi)
skewness_value = skew(tinggi)
# Tampilkan hasil ekstraksi fitur
print("Hasil Ekstraksi Fitur:")
print("Mean:", mean_value)
print("Variance:", variance_value)
print("Standard Deviation:", std_deviation_value)
print("Kurtosis:", kurtosis_value)
print("Skewness:", skewness_value)
```

# **Visualisasikan dalam bentuk grafik bar dan plot**

```python
# Tampilkan hasil ekstraksi fitur dalam bentuk bar
features = ['Mean', 'Variance', 'Std Deviation', 'Kurtosis',
'Skewness']
values = [mean_value, variance_value, std_deviation_value,
kurtosis_value, skewness_value]
plt.figure(figsize=(10, 6))
plt.bar(features, values, color='skyblue')
plt.title('Ekstraksi Fitur dan Tinggi Bangunan Maskot')
plt.xlabel('Fitur')
plt.ylabel('Nilai')
plt.show()
# Tampilkan tinggi bangunan maskot dalam bentuk bar
plt.figure(figsize=(10, 6))
plt.bar(maskot_bangunan, tinggi, color='green')
plt.title('Tinggi Bangunan Maskot')
plt.xlabel('Maskot Bangunan')
plt.ylabel('Tinggi (meter)')
plt.xticks(rotation=45, ha='right')
plt.show()
# Tampilkan tinggi bangunan maskot dalam bentuk bar dan hasil ekstraksi fitur dalam bentuk plot pada satu grafik
plt.figure(figsize=(12, 6))
# Bar plot untuk tinggi bangunan maskot
plt.bar(np.arange(len(maskot_bangunan)), tinggi, color='blue',
alpha=0.7, label='Tinggi Bangunan')
plt.ylabel('Tinggi (meter)')
plt.xticks(np.arange(len(maskot_bangunan)), maskot_bangunan,
rotation=45, ha='right')
# Plot untuk hasil ekstraksi fitur
plt.twinx()
plt.plot(np.arange(5), [mean_value, variance_value,
std_deviation_value, kurtosis_value, skewness_value], marker='o',
color='red', label='Ekstraksi Fitur')
plt.ylabel('Nilai Fitur')
plt.title('Tinggi Bangunan Maskot dan Hasil Ekstraksi Fitur')
plt.legend(loc='upper left')
plt.show()
```
