```python
!pip install wfdb
import wfdb
import matplotlib.pyplot as plt
import numpy as np
from scipy.fft import fft
```

```python
# Ganti sesuai folder Google Drive Anda
base_path = "/content/drive/MyDrive/prakpola/"
person_folder = "Person_04"
record_name = "rec_2"

# Gabungkan path lengkap
record_path = f"{base_path}/{person_folder}/{record_name}"

# Baca sinyal
record = wfdb.rdrecord(record_path)
signal = record.p_signal[:, 0]
fs = record.fs
t = np.linspace(0, len(signal)/fs, len(signal))
```

```python
plt.figure(figsize=(12, 4))
plt.plot(t, signal)
plt.title("Sinyal ECG - Domain Waktu")
plt.xlabel("Waktu (detik)")
plt.ylabel("Amplitudo (mV)")
plt.grid(True)
plt.show()
```

```python
amplitudo_max = np.max(signal)
energi = np.sum(signal ** 2)

print(f"Amplitudo Maksimum: {amplitudo_max:.4f} mV")
print(f"Energi Sinyal: {energi:.2f}")
```

```python
frekuensi = np.fft.fftfreq(len(signal), d=1/fs)
spectrum = np.abs(fft(signal))

plt.figure(figsize=(12, 4))
plt.plot(frekuensi[:len(frekuensi)//2], spectrum[:len(spectrum)//2])
plt.title("Spektrum Frekuensi Sinyal ECG (FFT)")
plt.xlabel("Frekuensi (Hz)")
plt.ylabel("Magnitudo")
plt.grid(True)
plt.show()
```

# **Grafik Nilai Sinyal ECG**

```python
plt.figure(figsize=(12, 4))
plt.plot(t, signal, color='yellow')
plt.title("Grafik Sinyal ECG - Domain Waktu")
plt.xlabel("Waktu (detik)")
plt.ylabel("Amplitudo")
plt.grid(True)
plt.show()
```

# **ANALISIS**
Grafik Sinyal ECG
Domain Waktu

Grafik ini menampilkan sinyal ECG Anda dalam domain waktu, menunjukkan bagaimana kekuatan sinyal (amplitudo) berubah dari waktu ke waktu. Ini memberikan gambaran visual dari bentuk gelombang khas detak jantung, yang terdiri dari gelombang P, kompleks QRS, dan gelombang T. Sumbu X menunjukkan waktu dalam detik, dan sumbu Y menunjukkan amplitudo dalam milivolt. Grafik ini penting untuk melihat pola detak jantung dan mengidentifikasi anomali visual.

# **Grafik Spektrum Frekuensi Sinyal ECG (FFT).**

```python
frekuensi = np.fft.fftfreq(len(signal), d=1/fs)
spectrum = np.abs(fft(signal))

plt.figure(figsize=(12, 4))
plt.plot(frekuensi[:len(frekuensi)//2], spectrum[:len(spectrum)//2], color='blue')
plt.title("Grafik Spektrum Frekuensi Sinyal ECG (FFT)")
plt.xlabel("Frekuensi (Hz)")
plt.ylabel("Magnitudo")
plt.grid(True)
plt.show()
```

# **ANALISIS**
Grafik Spektrum Frekuensi Sinyal ECG (FFT):

Grafik ini adalah representasi sinyal ECG dalam domain frekuensi. Dengan menerapkan Fast Fourier Transform (FFT), kita dapat melihat komponen frekuensi yang menyusun sinyal asli. Sumbu X menunjukkan frekuensi dalam Hertz (Hz), dan sumbu Y menunjukkan kekuatan (magnitudo) dari setiap frekuensi. Puncak-puncak pada grafik ini menyoroti frekuensi yang paling dominan dalam sinyal. Analisis frekuensi ini berguna untuk memahami karakteristik sinyal yang tidak mudah dilihat di domain waktu, seperti mendeteksi keberadaan noise pada frekuensi tertentu.
