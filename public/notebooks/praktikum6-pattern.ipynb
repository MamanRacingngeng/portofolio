```python
import pandas as pd
from ucimlrepo import fetch_ucirepo
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score
import plotly.express as px

# 1. Ambil dataset
#wine = fetch_ucirepo(id=109)
#X = wine.data.features
#print(X.head())

# Dataset postest
student_performance = fetch_ucirepo(id=320)
X = student_performance.data.features
print(X.head())
```

POSTEST

```python
# 2. Normalisasi
scaler = StandardScaler()

# Identifikasi kolom non-numerik
non_numeric_cols = X.select_dtypes(exclude=['number']).columns

# Buat DataFrame baru hanya dengan kolom numerik
X_numeric = X.drop(columns=non_numeric_cols)

# Terapkan scaler pada data numerik
X_scaled = scaler.fit_transform(X_numeric)

# 3. PCA ke 4 dimensi
# Perhatikan: PCA sekarang akan dilakukan pada X_scaled yang hanya berisi data numerik
pca = PCA(n_components=4)
X_pca = pca.fit_transform(X_scaled)

# 4. K-Means dengan 4 klaster
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
kmeans_labels = kmeans.fit_predict(X_scaled)
kmeans_score = silhouette_score(X_scaled, kmeans_labels)

# 5. Agglomerative Clustering dengan 4 klaster
agglo = AgglomerativeClustering(n_clusters=4)
agglo_labels = agglo.fit_predict(X_scaled)
agglo_score = silhouette_score(X_scaled, agglo_labels)

# 6. Buat DataFrame gabungan untuk plot (menggunakan 2 komponen PCA pertama untuk visualisasi 2D)
# Gunakan data PCA yang sudah dihitung dari data numerik
df_plot = pd.DataFrame({
    'PC1': X_pca[:, 0],
    'PC2': X_pca[:, 1],
    'KMeans Cluster': kmeans_labels.astype(str),
    'Agglo Cluster': agglo_labels.astype(str)
})

# 7. Plot interaktif dengan Plotly (menggunakan PC1 dan PC2 untuk visualisasi 2D)
fig1 = px.scatter(df_plot, x='PC1', y='PC2', color='KMeans Cluster',
                  title=f"K-Means Clustering (4 Clusters, Silhouette Score: {kmeans_score:.2f})",
                  labels={'color': 'Cluster'},
                  width=600, height=400)
fig2 = px.scatter(df_plot, x='PC1', y='PC2', color='Agglo Cluster',
                  title=f"Agglomerative Clustering (4 Clusters, Silhouette Score: {agglo_score:.2f})",
                  labels={'color': 'Cluster'},
                  width=600, height=400)

# 8. Tampilkan plot
fig1.show()
fig2.show()
```

PENJELASAN POSTEST

Penelitian ini menggunakan data kinerja siswa untuk mengelompokkan siswa ke dalam 4 grup menggunakan dua metode: K-Means dan Agglomerative Clustering.

1. Data siswa diambil dan disiapkan dengan menormalisasi fitur numerik dan mengurangi dimensinya menggunakan PCA.
2. Kedua algoritma clustering diterapkan pada data yang sudah diproses.
Hasil clustering dievaluasi menggunakan Silhouette Score.
3. Visualisasi dibuat menggunakan plot scatter berdasarkan hasil PCA untuk menampilkan cluster-cluster yang terbentuk oleh K-Means dan Agglomerative Clustering.

Penjelasan Visualisasi

Kedua visualisasi ini menampilkan hasil pengelompokan siswa ke dalam 4 grup menggunakan metode K-Means dan Agglomerative Clustering, setelah data siswa disederhanakan menjadi dua dimensi (PC1 dan PC2) menggunakan PCA; setiap titik mewakili siswa dengan warna yang menunjukkan kelompoknya, dan judul grafik menyertakan Silhouette Score sebagai ukuran kualitas pengelompokan, memungkinkan perbandingan visual antara hasil kedua metode clustering tersebut.

Perbandingan

Kedua visualisasi yang disajikan memberikan representasi visual dari hasil pengelompokan (clustering) siswa menggunakan dua algoritma yang berbeda: K-Means dan Agglomerative Clustering. Warna pada setiap titik mewakili kelompok tempat siswa diklasifikasikan oleh masing-masing algoritma. Perbandingan visual antara kedua plot memungkinkan kita untuk melihat bagaimana kedua metode tersebut membagi data menjadi empat kelompok yang berbeda secara spasial.

Meskipun keduanya bertujuan untuk mengelompokkan data, perbedaan dalam cara K-Means (berbasis pusat klaster) dan Agglomerative Clustering (berbasis hirarki) beroperasi tercermin dalam sebaran titik-titik dalam setiap kelompok. Plot juga menyertakan Silhouette Score untuk setiap metode, memberikan metrik kuantitatif mengenai kualitas pengelompokan. Dengan membandingkan Silhouette Score dan pola visual dari sebaran klaster, kita dapat mengevaluasi efektivitas relatif dari K-Means dan Agglomerative Clustering dalam mengelompokkan data siswa berdasarkan fitur-fitur yang digunakan.
